```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
#loading training data 
training_data_df = pd.read_csv("sales_data_training.csv", dtype=float)
x_training = training_data_df.drop('total_earnings', axis=1).values
y_training = training_data_df[['total_earnings']].values
```

```python
#load testing data set
testing_data_df = pd.read_csv("sales_data_test.csv" ,dtype=float)
x_testing = testing_data_df.drop('total_earnings', axis=1).values
y_testing = testing_data_df[['total_earnings']].values
```

```python
#scaling training Data
X_Scaler = MinMaxScaler(feature_range=(0,1))
Y_Scaler = MinMaxScaler(feature_range=(0,1))
```

```python
X_Scaled_training = X_Scaler.fit_transform(x_training)
Y_Scaled_training = Y_Scaler.fit_transform(y_training)
```

```python
#scaling testing Data
X_Scaled_testing = X_Scaler.fit_transform(x_testing)
Y_Scaled_testing = Y_Scaler.fit_transform(y_testing)
```

```python
#defining Model Parameters
learning_rate=0.001
training_epochs=100
display_step=5
#defining number of input and output
number_of_input = 9 
number_of_output = 1
#defining how many neurons we want in each step
layer_1_nodes = 50
layer_2_nodes = 100
layer_3_nodes = 50 
```

```python
tf.compat.v1.disable_eager_execution()


#modelling the layer 
with tf.compat.v1.variable_scope('input'):
    X = tf.compat.v1.placeholder(tf.float32, shape=(None, number_of_input))

with tf.compat.v1.variable_scope('layer_1'):
    weights=tf.compat.v1.get_variable(name='weights1', shape=[number_of_input, layer_1_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases=tf.compat.v1.get_variable(name='biases_1', shape = [layer_1_nodes], initializer=tf.zeros_initializer())
    layer_1_output=tf.nn.relu(tf.matmul(X,weights)+biases) #ax+b
    

with tf.compat.v1.variable_scope('layer_2'):
    weights=tf.compat.v1.get_variable(name='weights2', shape=[layer_1_nodes, layer_2_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases=tf.compat.v1.get_variable(name='biases_2', shape = [layer_2_nodes], initializer=tf.zeros_initializer())
    layer_2_output=tf.nn.relu(tf.matmul(layer_1_output,weights)+biases) #ax+b

with tf.compat.v1.variable_scope('layer_3'):
    weights=tf.compat.v1.get_variable(name='weights3', shape=[layer_2_nodes, layer_3_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases=tf.compat.v1.get_variable(name='biases_3', shape = [layer_3_nodes], initializer=tf.zeros_initializer())
    layer_3_output=tf.nn.relu(tf.matmul(layer_2_output,weights)+biases) #ax+b
    
with tf.compat.v1.variable_scope('output'):
    weights=tf.compat.v1.get_variable(name='weights4', shape=[layer_3_nodes, number_of_output], initializer=tf.keras.initializers.GlorotNormal())
    biases=tf.compat.v1.get_variable(name='biases_4', shape = [number_of_output], initializer=tf.zeros_initializer())
    prediction=tf.matmul(layer_3_output,weights)+biases #ax+b
    
```

```python
# Input Layer
with tf.compat.v1.variable_scope('input'):
    X = tf.compat.v1.placeholder(tf.float32, shape=(None, number_of_input))

# Layer 1
with tf.compat.v1.variable_scope('layer_1'):
    weights = tf.compat.v1.get_variable(name="weights1", shape=[number_of_input, layer_1_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases = tf.compat.v1.get_variable(name="biases1", shape=[layer_1_nodes], initializer=tf.zeros_initializer())
    layer_1_output = tf.nn.relu(tf.matmul(X, weights) + biases)

# Layer 2
with tf.compat.v1.variable_scope('layer_2'):
    weights = tf.compat.v1.get_variable(name="weights2", shape=[layer_1_nodes, layer_2_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases = tf.compat.v1.get_variable(name="biases2", shape=[layer_2_nodes], initializer=tf.zeros_initializer())
    layer_2_output = tf.nn.relu(tf.matmul(layer_1_output, weights) + biases)

# Layer 3
with tf.compat.v1.variable_scope('layer_3'):
    weights = tf.compat.v1.get_variable(name="weights3", shape=[layer_2_nodes, layer_3_nodes], initializer=tf.keras.initializers.GlorotNormal())
    biases = tf.compat.v1.get_variable(name="biases3", shape=[layer_3_nodes], initializer=tf.zeros_initializer())
    layer_3_output = tf.nn.relu(tf.matmul(layer_2_output, weights) + biases)

# Output Layer
with tf.compat.v1.variable_scope('output'):
    weights = tf.compat.v1.get_variable(name="weights4", shape=[layer_3_nodes, number_of_output], initializer=tf.keras.initializers.GlorotNormal())
    biases = tf.compat.v1.get_variable(name="biases4", shape=[number_of_output], initializer=tf.zeros_initializer())
    prediction = tf.matmul(layer_3_output, weights) + biases
```
